```python
# Initialize the environment
env = gym.make("LunarLander-v3", render_mode = 'human')

# reset env to generate the first observation
observation, info = env.reset(seed=42)
print(observation)
for _ in range(10):
  action = env.action_space.sample()
  observation, reward, terminate, truncated, info = env.step(action)
  print(observation)
  if terminate or truncated:
    observation, info = env.reset()

env.close()
```

## Basic

```python
# try with random action
env = gym.make("CartPole-v1", render_mode="human")

# reset to start a new episode
observation, info = env.reset()
print(f"Starting observation: {observation}")
episode_over = False
total_reward = 0

if not episode_over:
  action = env.action_space.sample()

  observation, reward, terminate, truncated, info = env.step(action)
  total_reward += reward
  episode_over = terminate or truncated

print(f"Episode over! Total reward: {total_reward}")
env.close()
```
